# Phase 6: Feature Engineering

## Objective

Feature engineering transforms the integrated retail dataset into a machine-learning and analytics-ready dataset.

In this notebook, we create:

1. Time-based features
2. Order and delivery features
3. Product features
4. Customer behavior features
5. RFM customer segmentation features
6. Sales features
7. Machine learning encoding features

The final dataset will be saved as:

data/processed/final_feature_dataset.csv

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

pd.set_option(
    "display.max_columns",
    None
)

pd.set_option(
    "display.float_format",
    "{:.2f}".format
)
```

```python
df = pd.read_csv(
    "C:/Users/rames/notebook/DS-1/data/processed/master_retail_dataset.csv"
)
```

```python
df.head()
```

```python
date_columns = [

"order_purchase_timestamp",

"order_approved_at",

"order_delivered_carrier_date",

"order_delivered_customer_date",

"order_estimated_delivery_date"

]


for col in date_columns:

    df[col] = pd.to_datetime(df[col])
```

```python
df["purchase_year"] = (
    df["order_purchase_timestamp"]
    .dt.year
)
```

```python
df["purchase_month"] = (
    df["order_purchase_timestamp"]
    .dt.month
)
```

```python
df["purchase_month_name"] = (
    df["order_purchase_timestamp"]
    .dt.month_name()
)
```

```python
df["purchase_quarter"] = (
    df["order_purchase_timestamp"]
    .dt.quarter
)
```

```python
df["purchase_day"] = (
    df["order_purchase_timestamp"]
    .dt.day_name()
)
```

```python
df["is_weekend"] = (

df["order_purchase_timestamp"]
.dt.dayofweek >= 5

).astype(int)
```

```python
df["delivery_days"] = (

df["order_delivered_customer_date"]

-

df["order_purchase_timestamp"]

).dt.days
```

```python
df["delivery_delay_days"] = (

df["order_delivered_customer_date"]

-

df["order_estimated_delivery_date"]

).dt.days
```

```python
df["delivery_completed"] = np.where(

df["order_delivered_customer_date"].isnull(),

0,

1

)
```

```python
df["delivery_status"] = np.where(

df["delivery_completed"]==0,

"Not Delivered",

np.where(

df["delivery_delay_days"]<=0,

"On Time",

"Delayed"

)

)
```

```python
df["total_order_value"] = (

df["price"]

+

df["freight_value"]

)
```

```python
df["freight_percentage"] = np.where(

df["total_order_value"] > 0,

(df["freight_value"] /
df["total_order_value"]) * 100,

0

)
```

```python
df["product_volume_cm3"] = (

df["product_length_cm"]

*

df["product_height_cm"]

*

df["product_width_cm"]

)
```

```python
df["weight_category"] = pd.cut(

df["product_weight_g"],

bins=[
-1,
1000,
5000,
10000,
100000
],

labels=[
"Light",
"Medium",
"Heavy",
"Very Heavy"
]

)
```

```python
df["description_quality"] = pd.cut(

df["product_description_lenght"],

bins=[
-1,
100,
500,
10000
],

labels=[
"Unknown",
"Short",
"Long"
]

)
```

```python
customer_frequency = (

df.groupby(
"customer_unique_id"
)

["order_id"]

.nunique()

.reset_index()

)


customer_frequency.columns=[

"customer_unique_id",

"total_orders"

]
```

```python
df=df.merge(

customer_frequency,

on="customer_unique_id",

how="left"

)
```

```python
customer_spending=(

df.groupby(
"customer_unique_id"
)

["total_order_value"]

.sum()

.reset_index()

)


customer_spending.columns=[

"customer_unique_id",

"total_spending"

]
```

```python
df=df.merge(

customer_spending,

on="customer_unique_id",

how="left"

)
```

```python
df["customer_average_order_value"]=(

df["total_spending"]

/

df["total_orders"]

)
```

```python
reference_date = (
df["order_purchase_timestamp"]
.max()
)
```

```python
rfm = df.groupby(
"customer_unique_id"
).agg({

"order_purchase_timestamp":
lambda x:
(reference_date-x.max()).days,


"order_id":
"nunique",


"total_order_value":
"sum"

}).reset_index()
```

```python
rfm.columns=[

"customer_unique_id",

"Recency",

"Frequency",

"Monetary"

]
```

```python
df=df.merge(

rfm,

on="customer_unique_id",

how="left"

)
```

```python
df["price_category"] = pd.cut(

df["price"],

bins=[
-1,
50,
200,
500,
100000
],

labels=[
"Unknown",
"Low",
"Medium",
"High"
]

)
```

```python
df["order_status_encoded"]=(

df["order_status"]

.astype("category")

.cat.codes

)
```

```python
df["payment_type_encoded"]=(

df["payment_type"]

.astype("category")

.cat.codes

)
```

```python
df["product_category_name"] = (

df["product_category_name"]

.fillna("Unknown")

)


df["product_category_name_english"]=(

df["product_category_name_english"]

.fillna("Unknown")

)
```

```python
product_numeric_columns=[

"product_weight_g",

"product_height_cm",

"product_length_cm",

"product_width_cm",

"product_name_lenght",

"product_description_lenght",

"product_photos_qty"

]


for col in product_numeric_columns:

    df[col]=df[col].fillna(
        df[col].median()
    )
```

```python
df["price"] = df["price"].fillna(0)

df["freight_percentage"] = np.where(

df["total_order_value"] > 0,

(df["freight_value"] /
df["total_order_value"]) * 100,

0

)


df["total_order_value"] = (
    df["price"]
    +
    df["freight_value"]
)
```
