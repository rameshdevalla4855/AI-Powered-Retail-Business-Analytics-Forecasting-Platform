```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.float_format", "{:.2f}".format)
```

```python
customers = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_customers_dataset.csv")

geolocation = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_geolocation_dataset.csv")

order_items = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_order_items_dataset.csv")

payments = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_order_payments_dataset.csv")

reviews = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_order_reviews_dataset.csv")

orders = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_orders_dataset.csv")

products = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_products_dataset.csv")

sellers = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_sellers_dataset.csv")

translation = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/product_category_name_translation.csv")
```

```python
datasets = {

    "Customers": customers,

    "Geolocation": geolocation,

    "Orders": orders,

    "Order Items": order_items,

    "Payments": payments,

    "Reviews": reviews,

    "Products": products,

    "Sellers": sellers,

    "Category Translation": translation

}
```

```python
summary = []

for name, df in datasets.items():

    summary.append({

        "Dataset": name,

        "Rows": df.shape[0],

        "Columns": df.shape[1],

        "Missing Values": df.isnull().sum().sum(),

        "Duplicate Rows": df.duplicated().sum()

    })

summary_df = pd.DataFrame(summary)

summary_df
```

```python
for name, df in datasets.items():

    print("="*80)

    print(name)

    print("="*80)

    display(df.head())

    print()

    df.info()

    print("\n")
```

```python
for name, df in datasets.items():

    print("="*80)

    print(name)

    print("="*80)

    missing = (
    df.isnull()
      .sum()
      .sort_values(ascending=False)
)

missing = missing[missing > 0]

display(missing)
```

```python
missing_percent = (

df.isnull().mean()*100

).round(2)

missing_percent = missing_percent[missing_percent>0]

display(missing_percent)
```

```python
missing_summary=[]

for name,df in datasets.items():

    total=df.isnull().sum().sum()

    percent=round(
        total/(df.shape[0]*df.shape[1])*100,
        2
    )

    missing_summary.append({

        "Dataset":name,

        "Total Missing":total,

        "Missing %":percent

    })

missing_summary=pd.DataFrame(
    missing_summary
)

missing_summary
```

```python
duplicate_summary = []

for name, df in datasets.items():

    duplicate_summary.append({

        "Dataset": name,

        "Duplicate Rows": df.duplicated().sum()

    })

pd.DataFrame(duplicate_summary)
```
