```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns",None)
pd.set_option("display.float_format","{:.2f}".format)
```

```python
customers = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_customers_dataset.csv")

geolocation = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_geolocation_dataset.csv")

order_items = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_order_items_dataset.csv")

payments = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_order_payments_dataset.csv")

reviews = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_order_reviews_dataset.csv")

orders = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_orders_dataset.csv")

products = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_products_dataset.csv")

sellers = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/olist_sellers_dataset.csv")

translation = pd.read_csv("C:/Users/rames/notebook/DS-1/data/raw/product_category_name_translation.csv")
```

```python
datasets = {

    "Customers": customers,

    "Geolocation": geolocation,

    "Orders": orders,

    "Order Items": order_items,

    "Payments": payments,

    "Reviews": reviews,

    "Products": products,

    "Sellers": sellers,

    "Translation": translation

}
```

```python
def data_summary(df,name):

    print("="*70)

    print(name)

    print("="*70)

    print("Shape :",df.shape)

    print("\nMissing Values")

    display(df.isnull().sum())

    print("\nDuplicate Rows :",df.duplicated().sum())

    print("\nData Types")

    display(df.dtypes)
```

```python
for name,df in datasets.items():

    data_summary(df,name)
```

```python
customers.drop_duplicates(inplace=True)
customers = customers.drop_duplicates().reset_index(drop=True)

print(customers.isnull().sum())

```

```python
customers.reset_index(drop=True,inplace=True)
```

```python
customers.info()
```

```python
geolocation.drop_duplicates(inplace=True)

print(geolocation.isnull().sum())

```

```python
geolocation.reset_index(drop=True,inplace=True)
```

```python
orders["order_purchase_timestamp"] = pd.to_datetime(
    orders["order_purchase_timestamp"]
)

orders["order_approved_at"] = pd.to_datetime(
    orders["order_approved_at"]
)

orders["order_delivered_carrier_date"] = pd.to_datetime(
    orders["order_delivered_carrier_date"]
)

orders["order_delivered_customer_date"] = pd.to_datetime(
    orders["order_delivered_customer_date"]
)

orders["order_estimated_delivery_date"] = pd.to_datetime(
    orders["order_estimated_delivery_date"]
)
```

```python
orders.drop_duplicates(inplace=True)
```

```python
orders["order_approved_at"] = orders["order_approved_at"].fillna(
    orders["order_purchase_timestamp"]
)
```

```python
orders["order_delivered_carrier_date"] = (
    orders["order_delivered_carrier_date"]
    .fillna(orders["order_estimated_delivery_date"])
)
```

```python
orders["order_delivered_customer_date"] = (
    orders["order_delivered_customer_date"]
    .fillna(orders["order_estimated_delivery_date"])
)
```

```python
orders.isnull().sum()
```

```python
order_items.drop_duplicates(inplace=True)
order_items["shipping_limit_date"] = pd.to_datetime(
    order_items["shipping_limit_date"]
)

```

```python
order_items["price"] = order_items["price"].fillna(
    order_items["price"].median()
)

order_items["freight_value"] = order_items["freight_value"].fillna(
    order_items["freight_value"].median()
)
```

```python
order_items.reset_index(drop=True,inplace=True)
```

```python
order_items.isnull().sum()
```

```python
payments.drop_duplicates(inplace=True)

payments["payment_type"] = payments["payment_type"].fillna("Unknown")

payments["payment_installments"] = (
    payments["payment_installments"]
    .fillna(0)
)

payments["payment_value"] = (
    payments["payment_value"]
    .fillna(
        payments["payment_value"].median()
    )
)

```

```python
payments.reset_index(drop=True,inplace=True)
```

```python
reviews["review_creation_date"]=pd.to_datetime(
reviews["review_creation_date"])

reviews["review_answer_timestamp"]=pd.to_datetime(
reviews["review_answer_timestamp"])
```

```python
reviews["review_comment_title"] = (
    reviews["review_comment_title"]
    .fillna("No Title")
)

reviews["review_comment_message"] = (
    reviews["review_comment_message"]
    .fillna("No Review")
)
```

```python
reviews.drop_duplicates(inplace=True)
```

```python
products.isnull().sum()
```

```python
products["product_category_name"].fillna(
"Unknown",
inplace=True
)
```

```python
products["product_name_lenght"].fillna(

products["product_name_lenght"].median(),

inplace=True

)
```

```python
products["product_description_lenght"].fillna(

products["product_description_lenght"].median(),

inplace=True

)
```

```python
products["product_photos_qty"].fillna(

products["product_photos_qty"].median(),

inplace=True

)
```

```python
products["product_weight_g"] = (
    products["product_weight_g"]
    .fillna(
        products["product_weight_g"].median()
    )
)
```

```python
products["product_length_cm"].fillna(

products["product_length_cm"].median(),

inplace=True

)
```

```python
products["product_height_cm"].fillna(

products["product_height_cm"].median(),

inplace=True

)
```

```python
products["product_width_cm"].fillna(

products["product_width_cm"].median(),

inplace=True

)
```

```python
products.drop_duplicates(inplace=True)
```

```python
sellers.drop_duplicates(inplace=True)


print(sellers.isnull().sum())
```

```python
sellers.reset_index(drop=True,inplace=True)
```

```python
translation.drop_duplicates(inplace=True)

translation.reset_index(drop=True,inplace=True)
```

```python
for name,df in datasets.items():

    print("="*70)

    print(name)

    print("="*70)

    print(df.isnull().sum())

    print()
```

```python
customers.to_csv("C:/Users/rames/notebook/DS-1/data/cleaned/customers_cleaned.csv", index=False)

geolocation.to_csv("C:/Users/rames/notebook/DS-1/data/cleaned/geolocation_cleaned.csv", index=False)

orders.to_csv("C:/Users/rames/notebook/DS-1/data/cleaned/orders_cleaned.csv", index=False)

order_items.to_csv("C:/Users/rames/notebook/DS-1/data/cleaned/order_items_cleaned.csv", index=False)

payments.to_csv("C:/Users/rames/notebook/DS-1/data/cleaned/payments_cleaned.csv", index=False)

reviews.to_csv("C:/Users/rames/notebook/DS-1/data/cleaned/reviews_cleaned.csv", index=False)

products.to_csv("C:/Users/rames/notebook/DS-1/data/cleaned/products_cleaned.csv", index=False)

sellers.to_csv("C:/Users/rames/notebook/DS-1/data/cleaned/sellers_cleaned.csv", index=False)

translation.to_csv("C:/Users/rames/notebook/DS-1/data/cleaned/translation_cleaned.csv", index=False)
```
